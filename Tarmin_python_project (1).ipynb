# **QUESTION 5**

```python
# QUESTION 5(a)

df_grouped_count = df_employee.groupby('Department').agg(Employee_Count=('EmployeeID', 'count')).sort_values(by='Employee_Count')
df_grouped_count
```

```python
# QUESTION 5(a)

import matplotlib.pyplot as plt

df_grouped_count.plot(kind = 'barh', title = 'Employee Distribution Across Departments', xlabel = 'Employee_Count', ylabel = 'Department',
                      figsize=(10, 6), color= 'steelblue', legend=False)

# Title fontsize
plt.title('Employee Distribution Across Departments', fontsize=14)

# Axis label fontsize
plt.xlabel('Employee Count', fontsize=12)
plt.ylabel('Department', fontsize=12)

# Add count labels on each bar
for index, value in enumerate(df_grouped_count['Employee_Count']):
    plt.text(value, index, str(value), va='center', fontsize=10)

# Prevents labels getting cut off
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# QUESTION 5(b)

# Cleaning of gender column in df_employee

df_employee['Gender']=df_employee['Gender'].replace({'Mle':'Male', 'Fmale':'Female'})
df_employee['Gender']
```

```python
# QUESTION 5(b)
df_grouped_gender = df_employee.groupby('Gender').agg(Employee_Count=('EmployeeID', 'count')).sort_values(by='Employee_Count')
df_grouped_gender
```

```python
import matplotlib.pyplot as plt

# QUESTION 5(b)

df_grouped_gender.plot(kind = 'bar', title = 'Employee Count By Gender', xlabel = 'Gender', ylabel = 'Employee_Count',
                      figsize=(8, 5), color= 'orange', legend=False)

# Title fontsize
plt.title('Employee Count By Gender', fontsize=14)

# Axis label fontsize
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Employee Count', fontsize=12)

# Add count labels on each bar
for index, value in enumerate(df_grouped_gender['Employee_Count']):
    plt.text(index, value, str(value), ha='center', fontsize=10)

# Prevents labels getting cut off
plt.tight_layout()

# keeps Gender labels horizontal
plt.xticks(rotation=0)

# Display the chart
plt.show()
```

```python
# QUESTION 5(c)

# Cleaning of gender column in df_health

df_health['Gender']=df_health['Gender'].replace({'Mle':'Male', 'Fmale':'Female'})
df_health['Gender']
```

```python
# QUESTION 5(c)

df_grouped_avg = df_health.groupby('Gender').agg(Average_StressLevel=('StressLevel','mean')).round(2).sort_values(by='Average_StressLevel')
df_grouped_avg
```

```python
# QUESTION 5(c)

df_grouped_avg.plot(kind = 'bar', title = 'Average Stress Level By Gender', xlabel = 'Gender', ylabel = 'Average_StressLevel',
                      figsize=(8, 5), color= 'green', legend=False)

# Title fontsize
plt.title('Average Stress Level By Gender', fontsize=14)

# Axis label fontsize
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Stress Level', fontsize=12)

# Add count labels on each bar
for index, value in enumerate(df_grouped_avg['Average_StressLevel']):
    plt.text(index, value, str(value), ha='center', fontsize=10)

# Prevents labels getting cut off
plt.tight_layout()

# keeps Gender labels horizontal
plt.xticks(rotation=0)

# Display the chart
plt.show()
```

```python
# QUESTION 5(d)

df_health.plot(kind='scatter',x='WorkHours', y = 'ProductivityScore', title = 'Productivity Score vs Work Hours', xlabel='Work Hours',
               ylabel='Productivity Score', figsize=(10, 6), color='red',alpha=0.6)

plt.tight_layout()
plt.show()

# ADD Correlation Score
correlation = df_health['WorkHours'].corr(df_health['ProductivityScore'])
print(f'Correlation between ProductivityScore and WorkHours: {round(correlation, 2)}')



```
